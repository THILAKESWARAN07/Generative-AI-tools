```python
category_count=df['category_canonical'].value_counts().head(5)

category_count.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 5 AI Tool Categories')
plt.ylabel('Number of Tools')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
modality_count=df['modality_canonical'].value_counts()
plt.figure(figsize=(6,4))
modality_count.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Modality')
plt.ylabel('Number of Tools')
plt.title('AI Modalities')
plt.grid(axis='y', color='gray', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
df = df.drop(columns=['api_status'])
```

```python
df['api_available'].value_counts()
```

```python
sns.countplot(x='api_available', data=df, palette='viridis')
plt.title('API Availability Distribution')
plt.xlabel('API Available (0=No, 1=Yes)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes('number').corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
df = df.sort_values(by='tool_name', ascending=True)
df = df.reset_index(drop=True)
```

```python
current_year = 2025
df['years_since_launch'] = current_year - df['release_year']
df[['tool_name', 'release_year', 'years_since_launch']].head(10)
```

```python
df['company_tool_counts'] = df.groupby('company')['tool_name'].transform('count')

df[['company', 'tool_name', 'company_tool_counts']].head(10)
```
